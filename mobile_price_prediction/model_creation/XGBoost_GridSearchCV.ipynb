```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
param_grid = {'max_depth': [3,4,5], 
              'learning_rate': [0.0001, 0.001, 0.01, 0.1, 1.0],
              'n_estimators': [50, 100, 500, 1000, 5000]}
grid = GridSearchCV(XGBRegressor(), param_grid, refit=True, verbose = 3, n_jobs = -1)
```

```python
grid.fit(independent_variable, dependent_variable)
```

```python
cv_results = grid.cv_results_
```

```python
grid.best_params_
```

```python
cv_results_df = pd.DataFrame.from_dict(cv_results)
```

```python
cv_results_df
```

```python
grid.best_score_
```

```python
result = grid.predict([[ppi, cpu_core, internal_mem, ram, cpu_freq,rearcam,thickness,battery]])
```
