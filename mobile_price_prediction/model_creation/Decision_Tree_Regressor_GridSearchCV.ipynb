```python
independent_variable = updated_dataset[features]
dependent_variable = updated_dataset[['price']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid = {'criterion': ['mse', 'mae', 'friedman_mse'],
             'max_features': ['auto', 'sqrt', 'log2'],
             'splitter': ['best', 'random']}
grid = GridSearchCV(DecisionTreeRegressor(), param_grid, refit=True, verbose =3, n_jobs=-1)
```

```python
grid.fit(independent_variable, dependent_variable)
```

```python
re = grid.cv_results_
```

```python
re_df = pd.DataFrame.from_dict(re)
```

```python
re_df
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python

ppi = int(input("ppi"))
cpu_core = int(input("cpu core:"))
internal_mem = float(input("internal mem:"))
ram = float(input("ram:"))
cpu_freq = float(input("cpu freq:"))
rearcam = float(input("rearcam:"))
thickness = float(input("thickness:"))
battery = int(input("battery:"))
```

```python
result = grid.predict([[ppi, cpu_core, internal_mem, ram, cpu_freq,rearcam,thickness,battery]])
```
