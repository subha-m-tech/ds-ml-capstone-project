```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
param_grid = {'n_estimators': [50, 100, 500, 1000, 5000], 
              'learning_rate': [0.0001,0.001,0.01, 0.1, 1.0],
             'loss': ['linear', 'square', 'exponential']}
grid = GridSearchCV(AdaBoostRegressor(), param_grid, refit=True, verbose =3, n_jobs =-1)
```

```python
grid.fit(independent_variable, dependent_variable)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
cv_results = grid.cv_results_
```

```python
re_df = pd.DataFrame.from_dict(cv_results)
```
