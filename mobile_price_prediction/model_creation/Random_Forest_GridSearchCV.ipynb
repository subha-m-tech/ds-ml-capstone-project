```python
import pandas as pd
```

```python
updated_dataset = pd.read_csv("../final_mob_dataset.csv")
```

```python
feature_list = pd.read_csv("../final_features.csv")
```

```python
features = feature_list.loc[:, "features_list"]
```

```python
updated_dataset[features]
```

```python
independent_variable = updated_dataset[features]
dependent_variable = updated_dataset[['price']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion': ['poisson', 'squared_error', 'friedman_mse', 'absolute_error'],
             'max_features': ['auto', 'sqrt', 'log2'],
             'n_estimators': [10,100]}
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)
```

```python
grid.fit(independent_variable, dependent_variable)
```

```python
grid.best_params_
```

```python
cv_results = grid.cv_results_
```

```python
cv_results_df = pd.DataFrame.from_dict(cv_results)
```

```python
cv_results_df
```

```python
grid.best_score_
```

```python
ppi = int(input("ppi"))
cpu_core = int(input("cpu core:"))
internal_mem = float(input("internal mem:"))
ram = float(input("ram:"))
cpu_freq = float(input("cpu freq:"))
rearcam = float(input("rearcam:"))
thickness = float(input("thickness:"))
battery = int(input("battery:"))
```

```python
result = grid.predict([[ppi, cpu_core, internal_mem, ram, cpu_freq,rearcam,thickness,battery]])
```
