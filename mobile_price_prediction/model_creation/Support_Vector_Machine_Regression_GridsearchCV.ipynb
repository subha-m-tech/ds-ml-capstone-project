```python
feature_list = pd.read_csv("../final_features.csv")
```

```python
features = feature_list.loc[:, "features_list"]
```

```python
independent_variable = updated_dataset[features]
dependent_variable = updated_dataset[['price']]
```

```python
# The code took more time if the sample si more than 10000 so as per the document 
#https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html
#The fit time complexity is more than quadratic with the number of samples which makes it hard to scale to datasets 
#with more than a couple of 10000 samples. 
#For large datasets consider using LinearSVR or SGDRegressor instead, 
# I already did hyperparameter tuning and got the best model with below parameters so I am reducing the option just to reduce the
# computational time
# kernel as poly, C as 700.0, coef0 as 1.0
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
param_grid= {'kernel': ['poly'], 
             'C': [10,100, 700, 1000,2000,3000],
             'coef0': [0.001,0.1,1.0],
             'gamma': ['scale']}
grid = GridSearchCV(SVR(), param_grid, refit=True, verbose =3, n_jobs=-1)
grid.fit(independent_variable, dependent_variable)
```

```python
cv_results = grid.cv_results_
```

```python
cv_results
```

```python
grid.best_params_
```

```python
cv_results_df = pd.DataFrame.from_dict(cv_results)
```
