```python
# Imports all the libraries
import pandas as pd
```

```python
# read the dataset
dataset1=pd.read_csv("../../final_mob_dataset.csv",index_col=None)
```

```python
# copy the dataset to df2
df2=dataset1
```

```python
# Print the dataset
df2
```

```python
# Convert the categorical data to numerical data using get_dummies
df2 = pd.get_dummies(df2, drop_first=True,dtype= int)
```

```python
# print the updated data
df2
```

```python
# Get independent and dependent variables
indep_X=df2.drop(columns='price', axis=1)
dep_Y=df2['price']
```

```python
# Print the indep_X
indep_X
```

```python
# Print the dep_Y
dep_Y
```

```python
# import regression related methods from feature_importance_regression
from feature_importance_regression import *
```

```python
# get best features
ensemble_feature_list=permutation_feature_importance(indep_X,dep_Y,8, df2.columns)
```

```python
# print selected features
ensemble_feature_list
```

```python
# Create empty list to store accuracy for each algorithms
acclin=[]
accsvml=[]
accsvmnl=[]
accdes=[]
accrf=[]
```

```python
# Get all the required details from each algorithm and assign the R2 score to the respective list
  
X_train, X_test, y_train, y_test=split_scalar(indep_X[ensemble_feature_list],dep_Y)  
r2_lin=Linear(X_train,y_train,X_test,y_test)
acclin.append(r2_lin)

r2_sl=svm_linear(X_train,y_train,X_test,y_test)    
accsvml.append(r2_sl)

r2_NL=svm_NL(X_train,y_train,X_test,y_test)
accsvmnl.append(r2_NL)

r2_d=Decision(X_train,y_train,X_test,y_test)
accdes.append(r2_d)

r2_r=random_forest(X_train,y_train,X_test,y_test)
accrf.append(r2_r)
```

```python
# get all the accuracy of each algorithm and print it in a dataframe
result=feature_importance_regression(acclin,accsvml,accsvmnl,accdes,accrf)
```
